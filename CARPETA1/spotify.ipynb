Esto simplemente nos confirma lo que anteriormente se había mencionado, ahora queremos analizar el top 10 de artistas, álbumes, canciones y géneros más populares.

````python
print("\nArtistas más frecuentes (Top 10)")
print(sputy['artists'].value_counts().head(10))
````

En primer lugar aparece The Beatles, seguido de manera muy cercana por George Jones y después Stevie Wonder.

````python
print("\nÁlbumes más frecuentes (Top 10)")
print(sputy['album_name'].value_counts().head(10))
````

El álbum más popular es Alternative Christmas 2022, seguido de Feliz Cumpleaños con Perreo y en tercer lugar Metal.

````python
print("\nCanciones más frecuentes (Top 10)")
print(sputy['track_name'].value_counts().head(10))
````

Aquí quisiera mencionar que de las 10 canciones más populares, 4 de ellas se relacionan con el tema de la Navidad, estas canciones son: Run Rudolph Run (1), Frosty The Snowman (3), Christmas Time (6) y Sleigh Ride (8).

El número entre paréntesis indica el lugar en la lista.

````python
print("\nGéneros más frecuentes (Top 10)")
print(sputy['track_genre'].value_counts().head(10))
````

Al analizar esta sección ocurre algo interesante, dado que los 10 géneros con más observaciones tienen la misma cantidad de registros, que son 1000, y ya que previamente se mencionó que el número total de géneros eran 114, podemos inferir que todos los géneros que se presentan tendrán un acumulado total de 1000 registros cada uno, generando un total final de 114,000.
